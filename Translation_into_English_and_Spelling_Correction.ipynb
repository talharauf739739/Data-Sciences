**Importing Libraries**

```python
!pip install googletrans==4.0.0-rc1
import pandas as pd
from googletrans import Translator

# Read the CSV file with the specified encoding
data = pd.read_csv('/content/QA_BOT_Performance.csv', encoding='latin1')

# Initialize Google Translate API
translator = Translator()

# Function to translate Roman Urdu to English
def translate_roman_urdu_to_english(text):
    translated_text = translator.translate(text, src='ur', dest='en')
    return translated_text.text

# Apply translation function to each row in the DataFrame
data['Translation'] = data['question'].apply(translate_roman_urdu_to_english)

# Print the DataFrame with translations
print(data)
```

```python
data.to_excel('check_translation.xlsx', index = False)
```

**Spelling Correction**

```python
import pandas as pd
from textblob import TextBlob


# Read the Excel file
"""data_frame = pd.read_excel('/content/Spelling_ko_thek_kro.xlsx')
# Specify the column containing addresses
address_column = 'address'"""



# Function to apply spell correction to an address
def correct_spelling(spelling):
    blob = TextBlob(spelling)
    corrected_spelling = str(blob.correct())
    return corrected_spelling



# Apply spell correction to addresses
data['spelling_correction'] = data['question'].apply(correct_spelling)


# Save the corrected data back to the Excel file
output_excel_file = 'check_correction_of_spelling.xlsx'
data.to_excel(output_excel_file, index=False)

```
