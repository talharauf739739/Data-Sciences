```python
import pandas as pd
import numpy as np
import regex as re
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/podump-2023-11-06.csv')
data.head()
```

```python
data = data[(data['Courier'] == 'selfpickup-safa') | (data['Courier'] == 'selfpickup-ufone')]
data
```

**Group By Phone Numbers**

```python
Persons_Repeated_phone = data.groupby(['Phone'])['Product Category'].count().reset_index()
#Persons_Repeated_phone.columns = ['Consignee Name', 'Coupon Count']
Persons_Repeated_phone
```

```python
phone = data[data['Phone'].duplicated(keep=False)]

# Display or work with the duplicates DataFrame
print(phone)

phone.to_excel('/content/phone_counts.xlsx', index =  False)
```

```python
Phone_against_product = phone.groupby(['Phone'])['Product Title'].count().reset_index()
#Persons_Repeated_phone.columns = ['Consignee Name', 'Coupon Count']
Phone_against_product.to_excel('Phone_against_product.xlsx', index = False)
```

**Now, Extract Consignee Names and Email, By Comparing phone_counts with Phone_against_product**

```python
# Merge dataframes based on the 'Phone' column
phone_counts = pd.read_excel('/content/phone_counts.xlsx')
merged_data = phone_counts.merge(Phone_against_product, on='Phone', how='inner')

# Extract Consignee Names and Email from the merged dataframe
result = merged_data[['Consignee Name', 'Email', 'Phone']]


result = result.drop_duplicates()
print(result)
result.to_excel('names_emails_for_duplicates.xlsx', index = False)
```

**Now, Aliegn data against phone number**

```python
data_1 = pd.read_excel('/content/data_1.xlsx')
print(data_1.head())
data_2 = pd.read_excel('/content/data_2.xlsx')


# Create a new column 'Count' in data_2
data_2['Count'] = None  # Initialize the 'Count' column with None

# Loop through the 'Phone' column in data_1 and data_2
for phone in data_2['Phone']:
    matching_product = data_1.loc[data_1['Phone'] == phone, 'Product Title'].values
    if matching_product:
        # Assign the matching 'Product_title' to 'Count' in data_2
        data_2.loc[data_2['Phone'] == phone, 'Count'] = matching_product[0]

print(data_2)
data_2.to_excel('llll.xlsx', index = False)
```
